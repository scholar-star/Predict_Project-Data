```python
address_data = pd.read_csv('D:\\Predict_Project Data\\국가철도공단_부산_지하철_주소데이터_20241021 (1).csv', encoding='cp949')
address_data
```

```python
def addressing(row):
    if address_data[address_data['역명'].str.contains(row['역명'])].size != 0:
        return address_data[address_data['역명'].str.contains(row['역명'])]['지번주소'].iloc[0]
    
subway_data['지번주소'] = subway_data.apply(addressing, axis=1)
subway_data
```

```python
subway_data.isnull().sum()
```

```python
def distinct(row):
    print(row['지번주소'].split(' ')[0])
    city = row['지번주소'].split(' ')[0]
    if city != "부산광역시":
        return "양산상북"
    else:
        dist = row['지번주소'].split(' ')[1]
        if dist == "사상구":
            return "사상"
        if dist == "강서구":
            return "가덕도"
        if dist == "북구":
            return "북구"
        if dist == "동래구":
            return "동래"
        if dist == "사하구":
            return "사하"
        if dist == "기장군":
            return "기장"
        if dist == "해운대구":
            return "해운대"
        if dist == "금정구":
            return "금정구"
        if dist == "남구":
            return "부산남구"
        else:
            return "부산"
        
subway_data['기상구역'] = subway_data.apply(distinct, axis=1)
        
```

```python
# 시간대별 열 목록 정의
time_columns = ['01시-02시', '02시-03시', '03시-04시', '04시-05시', '05시-06시', '06시-07시', 
                '07시-08시', '08시-09시', '09시-10시', '10시-11시', '11시-12시', '12시-13시', 
                '13시-14시', '14시-15시', '15시-16시', '16시-17시', '17시-18시', '18시-19시', 
                '19시-20시', '20시-21시', '21시-22시', '22시-23시', '23시-24시', '24시-01시']

# 시간대별 열을 제외한 기본 열들
id_columns = ['역번호', '역명', '년월일', '요일', '구분', '지번주소', '기상구역']

# melt 함수를 사용하여 시간대별 열을 행으로 변환
subway_data_melted = pd.melt(subway_data, 
                            id_vars=id_columns,
                            value_vars=time_columns,
                            var_name='시간',
                            value_name='승하차인원')

print("변환 후 데이터 형태:")
print(subway_data_melted.shape)
print("\n처음 10행:")
subway_data_melted.head(10)
subway_data_melted.tail(10)
```
