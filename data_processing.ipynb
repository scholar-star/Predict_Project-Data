```python
def addressing(row):
    if address_data[address_data['역명'] == row['역명']].size != 0:
        print(address_data[row['역명']==address_data['역명']]['지번주소'].iloc[0])
        return address_data[row['역명']==address_data['역명']]['지번주소'].iloc[0]
    
subway_data['지번주소'] = subway_data.apply(addressing, axis=1)
subway_data
```

```python
def distinct(row):
    city = row['지번주소'].split(' ')[0]
    if city != "부산광역시":
        return "양산상북"
    else:
        dist = row['지번주소'].split(' ')[1]
        if dist == "사상구":
            return "사상"
        if dist == "강서구":
            return "가덕도"
        if dist == "북구":
            return "북구"
        if dist == "동래구":
            return "동래"
        if dist == "사하구":
            return "사하"
        if dist == "기장군":
            return "기장"
        if dist == "해운대구":
            return "해운대"
        if dist == "금정구":
            return "금정구"
        if dist == "남구":
            return "부산남구"
        else:
            return "부산"
        
subway_data['구분'] = subway_data.apply(distinct, axis=1)
        
```
